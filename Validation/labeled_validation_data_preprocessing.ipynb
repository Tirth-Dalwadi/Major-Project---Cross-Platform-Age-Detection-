```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text_with_stop_words(text):
    # Lowercasing: Convert text to lowercase
    text = text.lower()

    # Remove usernames
    text = re.sub(r'@(\w+)', '', text)

    # Tokenization: Split the text into words
    words = nltk.word_tokenize(text)

    # Remove punctuation and non-alphanumeric characters
    words = [word for word in words if re.match(r'^[a-z0-9]+$', word)]

    # Remove English stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Join the words back into a single string
    preprocessed_text = ' '.join(words)

    return preprocessed_text

def preprocess_text_without_stop_words(text):
    # Lowercasing: Convert text to lowercase
    text = text.lower()

    # Remove usernames
    text = re.sub(r'@(\w+)', '', text)

    # Tokenization: Split the text into words
    words = nltk.word_tokenize(text)

    # Remove punctuation and non-alphanumeric characters
    words = [word for word in words if re.match(r'^[a-z0-9]+$', word)]

    # Join the words back into a single string
    preprocessed_text = ' '.join(words)

    return preprocessed_text
```

```python
df['cleaned_text_1'] = df['text'].apply(preprocess_text_with_stop_words)
df['cleaned_text_2'] = df['text'].apply(preprocess_text_without_stop_words)
```
