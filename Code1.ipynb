```python
data = pd.read_csv('/content/drive/MyDrive/Major_Project/Codebase/training/unlabeled_preprocessed_training.csv', encoding='utf-8')
data
```

```python
null_count = data['cleaned_text_2'].isnull().sum()
null_count
```

```python
column_name = 'cleaned_text_2'  # Specify the column name
data = data.dropna(subset=[column_name])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

selected_df = data[['cleaned_text_2', 'id']]

# Initialize CountVectorizer for bag of ngrams
ngram_vectorizer = CountVectorizer()  # You can adjust ngram_range as needed

# Select only the first 10000 rows of the 'cleaned_text_2' column
selected_df_subset = selected_df['cleaned_text_2'][:10000]

# Fit and transform the text data for the subset
X_ngrams_subset = ngram_vectorizer.fit_transform(selected_df_subset)

# Convert the sparse matrix to a DataFrame for the subset
ngram_df_subset = pd.DataFrame(X_ngrams_subset.toarray(), columns=ngram_vectorizer.get_feature_names_out())

ngram_df_subset
```

```python
ngram_df_subset.to_csv('/content/drive/MyDrive/Major_Project/Codebase/training/vectorized_unlabeled.csv', index=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

selected_df = data[['cleaned_text_2', 'id']]

# Initialize CountVectorizer for bag of ngrams
ngram_vectorizer = CountVectorizer()  # You can adjust ngram_range as needed

# Select only the first 10000 rows of the 'cleaned_text_2' column
selected_df_subset = selected_df['cleaned_text_2'][:20000]

# Fit and transform the text data for the subset
X_ngrams_subset = ngram_vectorizer.fit_transform(selected_df_subset)

# Convert the sparse matrix to a DataFrame for the subset
ngram_df_subset = pd.DataFrame(X_ngrams_subset.toarray(), columns=ngram_vectorizer.get_feature_names_out())

ngram_df_subset

# Append the subset DataFrame to the CSV file
ngram_df_subset.to_csv('/content/drive/MyDrive/Major_Project/Codebase/training/vectorized_unlabeled.csv', mode='a', index=False)
```
