```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
import pandas as pd
input = pd.read_csv('/content/drive/MyDrive/Major_Project/Codebase/training/labeled_preprocessed_training.csv', encoding='utf-8')
val = pd.read_csv('/content/drive/MyDrive/Major_Project/Codebase/validation/labeled_preprocessed_validation.csv', encoding='utf-8')
```

```python
val = val.dropna(subset="cleaned_text_1")
```

```python
input_text = input["cleaned_text_1"]
input_labels = input["label"]
val_text = val["cleaned_text_1"]
val_labels = val["label"]
```

```python
combined_text = input_text.append(val_text)
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
input_text_tfidf = tfidf_vectorizer.fit_transform(input_text)
val_text_tfidf = tfidf_vectorizer.transform(val_text)
```

```python
svm_model = SVC(kernel='linear')
```

```python
svm_model.fit(input_text_tfidf, input_labels)
```

```python
val_pred = svm_model.predict(val_text_tfidf)
```

```python
accuracy = accuracy_score(val_pred, val_labels)
```
